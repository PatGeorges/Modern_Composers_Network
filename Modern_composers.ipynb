#**Introduction** 

The **Common Practice Period** (1650–1900) of **Western classical music** is a **macrocosm** in that rules and practices were comparatively unified and composers could be described as belonging to comparatively long periods such as the late Renaissance, Baroque, Classical or Romantic. 

By contrast, the music of the twentieth century is both pluralist and **microcosmic** in that, throughout, composers explored more personal and individual approach to music creation, forming their own microcosms or ‘small universes’. 

This notebook uses Python libraries **Networkx** and **fa2/ForceAtlas2** to generate a network of 'modern' (mainly late 19th and first part of twentieth century) classical music composers. 

The assumption for the construction of this network is that if one composer has influenced another composer, there will be an edge linking these two composers (two nodes). 

The model is entirely explained in **Georges and Seckin (2022)**.  See reference at the end of the notebook.   

In this notebook (**Modern_composers.ipynb**) and the attached python script (**Modern_composers.py**) we share the code that permits to generate **Figures 5, 7 and 8 of Georges and Seckin (2022)**.

#**ForceAtlas2**

Although we use Networkx, we opted for the **ForecAtlas2 (fa2) algorithm** to generate the network graph per se. 

ForceAtlas2 algorithm (**Jacomy et al., 2014**), is a forced-based layout whereby the algorithm modifies an initial (random) node placement by continuously moving the nodes according to a system of forces based on a metaphor of springs and electric charged particles. The ‘spring-electric’ layout uses the attraction formula of springs (between nodes connected with an edge) and the repulsion formula of electrically charged particles (between any nodes).

For further details, see Jacomy, M., Venturini, T., Heymann, S., & Bastian, M. (2014). ForceAtlas2, a continuous graph layout algorithm for handy network visualisation designed for the Gephi software. PLoS ONE, 9(6), e98679.

The Python fa2 library is available here:

**Chippada, B. (2017). ForceAtlas2 for python.** https://github.com/bhargavchippada/forceatlas2. 

Some discussion surrounding the functioning of fa2 is also in Georges and Seckin (2022)
